```python
boston=pd.read_csv("C:/Users/User/Desktop/Boston.csv")
print(boston)
```

```python
#acessing more than one column in a dataset
boston[["tax","age","black"]]
```

```python
#filtering rows
boston[boston["age"]>50]
```

```python
#selecting specific rows and columns using loc
boston.loc[boston["age"]>50,"tax"]
```

```python
#selecting rows using iloc
boston.iloc[0:5]
```

```python
#slicing of data using iloc
boston.iloc[0:5,10:20]
```

```python
#selcecting figure in a particular row and column
boston.iloc[5,7]
```

```python
#doing the sum of a column
boston.tax.sum()
boston.sum()
```

```python
#doing the mean of a column
boston.mean(axis=1)
```

```python
boston.tax.mean()
```

```python
#it does the mean of the whole row
boston.mean(axis=1,skipna=False)
```

```python
#finding the maximum number of rows and column
boston.idxmax()
```
