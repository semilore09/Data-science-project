```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datax=[1,2,4,6,8,10]
datay=[0,3,6,9,12,15]
plt.plot(datax,datay)
plt.annotate(xy=[2,3],text="second point")#(it is used to label a graph)
plt.xlabel("Wheather")
plt.ylabel("Temperature")
plt.title("Wheather & Temperature Relationship")
plt.show()
```

```python
x=np.linspace(-4,4,num=100)
konstant=1/np.sqrt(2*np.pi)
konstant2=np.exp((-x**2)/2)
pdf=konstant*konstant2
plt.figure()
plt.plot(x,pdf,color="green")
plt.xlabel("random variable")
plt.ylabel("probability density")
plt.title("normal distribution")
plt.grid()
plt.show()
```

```python
#creating a stacked bar chart
x=["JAN","FEB","MAR","APR"]
y1=np.array([1000,2000,1000,3000])
y2=np.array([2000,2500,1500,2500])
y3=np.array([2500,3000,4500,5000])
plt.figure()
plt.bar(x,y1,color="black")
plt.bar(x,y2,bottom=y1,color="blue")
plt.bar(x,y3,bottom=y2+y1,color="cyan")
plt.xlabel("Months")
plt.ylabel("Cost")
plt.legend(["Food","Transport","Clothing"])
plt.savefig("C:/Users/User/Desktop/Data science/my images/Barchart.svg"edgwTrue)
plt.show()
```
