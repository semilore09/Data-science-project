```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ad=pd.read_csv("Housing Adaption Grant .csv")
print(ad)
```

```python
LA=ad["Local Authority"].iloc[0:11]
VOP=ad["Value of payments € 2014"].iloc[0:11]
plt.figure(figsize=[25,10])
sns.set_style("ticks")
sns.set_context("paper")
sns.barplot(x=LA,y=VOP,alpha=0.9,data=ad)
sns.palplot(sns.color_palette(palette="Greens"))
plt.show()
```

```python
VOP=ad["Value of payments € 2015"].iloc[0:11]
plt.figure()
sns.displot(x=VOP,data=ad,kind="kde")
plt.show()
```

```python
VOP=ad["Value of payments € 2015"].iloc[0:11]
plt.figure()
sns.displot(x=VOP,data=ad,hue=VOP)
plt.show()
```
