```python
fp=pd.read_csv("foodprice.csv")
print(fp)
```

```python
plt.figure(figsize=(10,5))
data1=fp.commodity.iloc[1:20].sort_values()
data2=fp.price.iloc[1:20].sort_values()
plt.plot(data1,data2,"r--")
plt.xlabel("commodity")
plt.ylabel("price")
plt.title("prices of food commodities in Nigeria")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
data1=fp.commodity.iloc[1:20]
data2=fp.price.iloc[1:20]
plt.bar(data1,data2)
plt.xlabel("commodity")
plt.ylabel("price")
plt.title("prices of food commodities in Naira(N)")


info1=fp.commodity.iloc[1:20]
info2=fp.usdprice.iloc[1:20]
plt.subplot(122)
plt.bar(info1,info2)
plt.xlabel("commodity")
plt.ylabel("usdprice")
plt.title("prices of food commodities in Usd($)")
plt.show()
```

```python
info1=fp.price.iloc[1:6]
info2=fp.commodity.iloc[1:6]
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(info2,info1,"k--")
plt.xlabel("commodity")
plt.ylabel("price")
plt.title("A graph of food price in Naira")

info=fp.usdprice.iloc[1:6]
info3=fp.commodity.iloc[1:6]
plt.subplot(122)
plt.plot(info3,info,"r--")
plt.xlabel("commodity")
plt.ylabel("usdprice")
plt.title("A graph of food price in usd")
plt.grid()
plt.legend("usd price")
plt.show()
```

```python
#bar charts-it is used for extensive 
```

```python
info1=fp.price.iloc[1:6].sort_values()
info2=fp.commodity.iloc[1:6].sort_values()
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.barh(info2,info1,color="black")
plt.xlabel("commodity")
plt.ylabel("price")
plt.legend("Naira price")
plt.title("A graph of food price in Naira")

info=fp.usdprice.iloc[1:6].sort_values()
info3=fp.commodity.iloc[1:6].sort_values()
plt.subplot(122)
plt.bar(info3,info,color="hotpink",width=0.6)
plt.xlabel("commodity")
plt.ylabel("usdprice")
plt.title("A graph of food price in usd")
plt.legend("usd price")
plt.show()
```

```python
#piechart is used for categorical data
#explode slice out a pie
myexplode=[0,0.2,0,0,0]
colours=["red","orange","yellow","blue","cyan","pink"]
info1=fp.price.iloc[1:6].sort_values()
plt.pie(info1,labels=fp.commodity.iloc[1:6].sort_values(),explode=myexplode,colors=colours,shadow=True)
plt.title("A graph of food price in Naira")
plt.startangle=90
plt.legend(title="commodities",loc="lower left",ncol=5)
```
